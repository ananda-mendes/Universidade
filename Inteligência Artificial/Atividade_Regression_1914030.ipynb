```python
#Importando as bibliotecas
import pandas as pd
import sklearn 
import math 
import numpy as np

import sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/drive/MyDrive/UFOP - Ananda/Sexto PeriÃÅodo/IA/candidatos.csv')
df
```

```python
# Preenchendo o valor ausente da coluna test_score com a m√©dia
df['test_score'] = df['test_score'].fillna(math.floor(df['test_score'].median()))
df
```

```python

# Mudando os atributos de str para int (tentei fazer com LabelEncoder por√©m os 
# n√∫meros adicionados foram de forma aleat√≥ria e preferi mudar com replace)

df['experience'] = df['experience'].replace("one", 1)
df['experience'] = df['experience'].replace("two", 2)
df['experience'] = df['experience'].replace("three", 3)
df['experience'] = df['experience'].replace("four", 4)
df['experience'] = df['experience'].replace("five", 5)
df['experience'] = df['experience'].replace("six", 6)
df['experience'] = df['experience'].replace("seven", 7)
df['experience'] = df['experience'].replace("eight", 8)
df['experience'] = df['experience'].replace("nine", 9)
df['experience'] = df['experience'].replace("ten", 10)
df['experience'] = df['experience'].replace("eleven", 11) 

# Preenchendo os valores ausentes com a m√©dia das experi√™ncias
df['experience'] = df['experience'].fillna(math.floor(df['experience'].median()))

df.astype(int)
```

```python
reg = linear_model.LinearRegression()

reg.fit(df[['experience', 'test_score', 'interview_score']], df['salary'])
```
