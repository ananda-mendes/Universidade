```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/drive/MyDrive/UFOP - Ananda/Sexto PeriÃÅodo/IA/decision_tree.csv')
df
```

```python
input = df.drop('salary', axis='columns')
target = df['salary']
input
```

```python
target
```

```python
# Conferindo os atributos categ√≥ricos
s = (input.dtypes == 'object')
categoricos = list(s[s].index)

print("Atributos categ√≥ricos:", categoricos)

oe = LabelEncoder()

# Aplicando o 'Label encoder' em cada coluna com atributos categ√≥ricos
for obj in categoricos:
  input[obj] = oe.fit_transform(df[obj].astype(object))

input
```

```python
model = tree.DecisionTreeClassifier()
model.fit(input, target)
```

```python
# abc_pharma = 0
# facebook = 1
# google = 2

# business manager = 0
# computer programmer = 1
# sales executive = 2

# bachelors = 0
# masters = 1

model.predict([[2, 2, 1]])
```
